```python
file_path = '../input/carpriceprediction/data.csv'
data_df = pd.read_csv(file_path)
data_df.head()
```

```python
# What's the average price of BMW cars in the dataset?
data_df[data_df['Make']=="BMW"].MSRP.mean()
```

```python
#Select a subset of cars after year 2015 (inclusive, i.e. 2015 and after). How many of them have missing values for Engine HP?
data_df[data_df['Year']>=2015]["Engine HP"].isnull().sum()
```

```python
# Question 5
mean_hp_before = data_df['Engine HP'].mean()
print(data_df['Engine HP'].isnull().sum())
data_df['Engine HP'].fillna(mean_hp_before)
mean_hp_after =data_df['Engine HP'].mean()
print(round(mean_hp_before))
print(round(mean_hp_after))
```

```python
# Question 6
data_roll = data_df[data_df['Make']== "Rolls-Royce"]
data_roll_filtered = data_roll[["Engine HP", "Engine Cylinders", "highway MPG"]]
data_roll_filtered = data_roll_filtered.drop_duplicates()
data_roll_filtered
```

```python
# get numpy array
x = data_roll_filtered.values
print(x)
```

```python
# Question 6
x_transpose = x.T
x_out = x_transpose.dot(x)
x_inv =  np.linalg.inv(x_out)
print(x_inv.sum())
```

```python
# Questions 7
y = [1000, 1100, 900, 1200, 1000, 850, 1300]
w = (x_inv.dot(x_transpose)).dot(y)
print(w)
```
